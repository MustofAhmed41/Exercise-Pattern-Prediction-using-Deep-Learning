## Deep Learning

```python
import tensorflow as tf
from tensorflow.keras import activations
from tensorflow.keras.utils import to_categorical
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.metrics import accuracy_score, mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
SEED = 33

import random

def reset_seeds(SEED):
   np.random.seed(SEED) 
   tf.random.set_seed(SEED)
   os.environ['PYTHONHASHSEED']=str(SEED)
   tf.random.set_seed(SEED)

random.seed(SEED)
reset_seeds(SEED) 
```

```python
df = load.copy()
```

```python
lst = []
for i in df['num_window'].unique():
    df.drop(df.loc[(df['new_window']=='no') & (df['num_window']==i)].index, inplace=True)
```

```python
df.shape
```

```python
def impute_classe(x):
  if x == 'A':
    return 4
  elif x =='B':
    return 3
  elif x=='C':
    return 2
  elif x=='D':
    return 1
  elif x=='E':
    return 0
```

```python
df['classe'] = df['classe'].apply(impute_classe)
```

```python
df['classe'].value_counts()
```

```python
lst = ['Unnamed: 0', 'classe', 'num_window', 'new_window']
X=df.drop(lst, axis=1)
Y=df['classe'] 
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1,stratify=Y)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.transform(X_test)
```

# Neural Network Model

```python
#from keras.layers import LeakyReLU
#model.add(LeakyReLU(alpha=0.05))

model=tf.keras.Sequential([
          tf.keras.layers.Dense(1200,activation='relu',input_shape=(X.shape[1],), use_bias=True,),        
          tf.keras.layers.Dropout(.2, input_shape=(1500,)),          
          tf.keras.layers.Dense(1500,activation='relu', use_bias=True,),  
          tf.keras.layers.Dropout(.2, input_shape=(1500,)), 
          tf.keras.layers.Dense(900,activation='relu', use_bias=True,),  
          tf.keras.layers.Dropout(.2, input_shape=(1000,)),  
          tf.keras.layers.Dense(700,activation='relu', use_bias=True,),  
          tf.keras.layers.Dense(350,activation='relu', use_bias=True,),  
          tf.keras.layers.Dense(5,activation='softmax')
])
```

```python
from tensorflow import keras
opt = keras.optimizers.Adam(learning_rate=0.00015)
model.compile(optimizer=opt, loss="sparse_categorical_crossentropy")
```

```python
model.summary()
```

```python
mean = []
for i in range(10):
  trained_model=model.fit(X_train,Y_train,epochs=50, batch_size=10 )
  Y_pred = model.predict(X_test)
  Y_pred = Y_pred.argmax(axis=1) 
  a = accuracy_score(Y_test, Y_pred)
  print('Accuracy for round : ', i ,' is ', a)
  mean.append(a)
```

```python
print('Mean Accuracy : ')
print(sum(mean)/len(mean))
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,Y_pred))
```
